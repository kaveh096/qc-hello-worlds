```python
# pip install networkx matplotlib
from pathlib import Path
import sys

# add repo root (parent of notebooks/) to sys.path
sys.path.append(str(Path.cwd().parent))
from lib.graphlib import GraphLib
```

```python
# Create and visualize a random graph
g = GraphLib()
g.generate_random_graph(num_vertices=12, edge_prob=0.4, seed=42)
g.visualize()
```

```python
# Max-Cut (Bruteforce vs Approximation)
cut_opt, weight_opt = g.max_cut_bruteforce()
cut_rand, weight_rand = g.max_cut_random(trials=1000)
cut_local, weight_local = g.max_cut_local_search()

print("Max-Cut Optimal:", cut_opt, weight_opt)
print("Max-Cut Random Approx:", cut_rand, weight_rand)
print("Max-Cut Local Search Approx:", cut_local, weight_local)
```

```python
# Create and visualize a random complete graph
g = GraphLib()
g.generate_random_graph(num_vertices=8, edge_prob=1.0, seed=42)
g.visualize()
```

```python
# TSP (Bruteforce vs Approximation)
if (not g.is_complete()):
    raise ValueError("Graph must be complete for TSP.")
path_opt, cost_opt = g.tsp_bruteforce()
path_nn, cost_nn = g.tsp_nearest_neighbor()
path_mst, cost_mst = g.tsp_mst_approximation()

print("TSP Optimal:", path_opt, cost_opt)
print("TSP Nearest Neighbor:", path_nn, cost_nn)
print("TSP MST Approximation:", path_mst, cost_mst)
```
